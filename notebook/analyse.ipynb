```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt



# Calculer le prix par m²
df['Price_per_m2'] = df['Price'] / df['Area']

# Agréger les prix moyens par m² par localisation et type de bien
price_per_m2_by_location_type = df.groupby(['Localisation', 'Type'])['Price_per_m2'].mean().unstack()

# Sélectionner les 10 localisations les plus chères
top_10_locations = price_per_m2_by_location_type.mean(axis=1).nlargest(10).index
price_per_m2_top_10 = price_per_m2_by_location_type.loc[top_10_locations]

# Sélectionner les 2 types de biens les plus fréquents (ou les plus pertinents)
top_2_types = price_per_m2_top_10.mean().nlargest(3).index
price_per_m2_top_10_top_2 = price_per_m2_top_10[top_2_types]

# Affichage des prix moyens par m² par localisation et type de bien pour les 10 localisations et 2 types de biens
plt.figure(figsize=(20, 80))
price_per_m2_top_10_top_2.plot(kind='bar')
plt.xlabel('Localisation')
plt.ylabel('Prix Moyen par m²')
plt.title('Prix Moyen par m² par Localisation et Type de Bien (Top 12 Localisations, Top 3 Types)')
plt.legend(title='Type de Bien')
plt.xticks(rotation=45)
# plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
import folium




# Initialisation du géocodeur
geolocator = Nominatim(user_agent="myGeocoder")

# Fonction pour obtenir les coordonnées géographiques
def get_coordinates(location):
    try:
        location = geolocator.geocode(location + ", Casablanca, Morocco")
        return (location.latitude, location.longitude)
    except:
        return None

# Ajout des coordonnées géographiques aux données
df['Coordinates'] = df['Localisation'].apply(get_coordinates)

# Filtrer les entrées sans coordonnées (au cas où)
df = df.dropna(subset=['Coordinates'])

# Créer une carte centrée sur Casablanca
map = folium.Map(location=[33.5731, -7.5898], zoom_start=12)

# Ajouter des cercles pour chaque bien immobilier
for index, row in df.iterrows():
    folium.CircleMarker(
        location=row['Coordinates'],
        radius=row['Price'] / 100000,  # Ajuster le diviseur pour la taille des cercles
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"Prix: {row['Price']}, Type: {row['Type']}"
    ).add_to(map)

# Sauvegarder la carte dans un fichier HTML
map.save("carte_biens_immobiliers.html")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, KBinsDiscretizer
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from imblearn.over_sampling import SMOTE



# Préparation des fonctionnalités et de la cible
features = ['Area', 'Rooms', 'Bedrooms', 'Bathrooms', 'Floor']
X = df[features]
y = df['Price']

# Standardisation des données
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Catégoriser les prix en classes
kbin = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
y_binned = kbin.fit_transform(y.values.reshape(-1, 1)).ravel()

# Appliquer SMOTE
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_scaled, y_binned)

# Séparation des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Création du modèle de régression
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Prédiction sur l'ensemble de test
y_pred = model.predict(X_test)

# Inverse transform pour obtenir les prix prédits originaux
y_pred_prices = kbin.inverse_transform(y_pred.reshape(-1, 1)).ravel()

# Évaluation du modèle
mse = mean_squared_error(y_test, y_pred_prices)
rmse = mse ** 0.5

print(f"Root Mean Squared Error (RMSE): {rmse}")

# Enregistrement du modèle si nécessaire
# import joblib
# joblib.dump(model, 'real_estate_price_model.pkl')
```
